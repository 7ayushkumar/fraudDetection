```python
import pandas as pd
import numpy as np
```

```python
fraud_testset = pd.read_csv('Insurance Fraud -TEST-12900(1).csv')
print(fraud_testset.shape)
fraud_testset.head()
```

```python
all_dummies = pd.get_dummies(fraud_testset, columns=['MONTH', 'DAYOFWEEK', 'MAKE', 'ACCIDENTAREA', 'DAYOFWEEKCLAIMED'
                                                     , 'MONTHCLAIMED', 'SEX', 'MARITALSTATUS', 'FAULT', 'POLICYTYPE'
                                                     , 'VEHICLECATEGORY', 'VEHICLECATEGORY', 'VEHICLEPRICE', 'DAYS_POLICY_ACCIDENT'
                                                     , 'DAYS_POLICY_CLAIM', 'PASTNUMBEROFCLAIMS', 'AGEOFVEHICLE'
                                                     , 'AGEOFPOLICYHOLDER', 'POLICEREPORTFILED', 'WITNESSPRESENT', 'AGENTTYPE'
                                                     , 'NUMBEROFSUPPLIMENTS', 'ADDRESSCHANGE_CLAIM', 'NUMBEROFCARS'
                                                     , 'YEAR', 'BASEPOLICY', 'FRAUDFOUND'], drop_first=True)
```

```python
fraud_testset = pd.concat([fraud_testset, all_dummies], axis=1)
```

```python
fraud_testset.to_csv("fraud_testset.csv",index=False)
```
